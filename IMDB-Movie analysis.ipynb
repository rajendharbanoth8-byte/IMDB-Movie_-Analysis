### 22. How many films were made in each genre?

```python
# Drop missing values
df_nonnull = df.dropna(subset=['Genre'])

# Normalize the Genre column: replace multiple separators and whitespace with a single comma
df_nonnull['Genre'] = df_nonnull['Genre'].str.replace(r'[\n\t]+', ' ', regex=True)  # remove newlines/tabs
df_nonnull['Genre'] = df_nonnull['Genre'].str.replace(r'\s*[,;/]\s*', ',', regex=True)  # unify separators

# Split genres and explode
all_genres = df_nonnull['Genre'].str.split(',').explode().str.strip()

# Remove empty strings
all_genres = all_genres[all_genres != '']

# Count films for each genre
genre_counts = all_genres.value_counts()

# Print each genre and its count
print("Number of films in each genre:")
for genre, count in genre_counts.items():
    print(f"{genre}: {count}")
```
